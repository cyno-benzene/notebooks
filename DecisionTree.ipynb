### Imported Libraries
Imported iris dataset, tree class which contains the desired DecisionTreeClassifier method, data splitting method and metrics method all from scikit learn.  

```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

### Declared dataset object
Created separate objects for iris data features and target labels.

```python
iris = load_iris()
X, y = iris.data, iris.target
```

### Splitting dataset into training and test
Splitted the dataset using the train_test_split() method. Three parameters features; target, and test_set size(30% of complete dataset) were passed in the method.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=80)
```

### Building Decision Tree Model

```python
clf = tree.DecisionTreeClassifier(random_state=0)
model = clf.fit(X_train, y_train)
```

### Testing the model
Tested the decision tree model using the test data.

```python
y_pred = model.predict(X_test)
```

### Evaluating the model
Made an estimation of how accurate the classifier model can predict the types of flowers. The result says the classification rate is 91% accurate.

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100,"%")
```

### Visualising the decision tree
By default skikit learn allows plotting the tree using its plot() method but it's output is very small. We can also use matplotlib or graphviz for better output display.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


plt.figure(figsize=(25, 10))
plot_tree(model,
          rounded=True,
          filled=True,
          proportion=True);
```
