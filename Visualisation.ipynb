```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(20)
y = np.arange(3,23)
print(x,"\n",y)
```

```python
plt.plot(x,y,  color='green', marker='o', linestyle='dotted', linewidth=2, markersize=10)
```
