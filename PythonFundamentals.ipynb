```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
len(df [ (df['sex'] == 'female' ) & (df['alive'] == 'yes' )  ]) # without using functions
```

```python
len(df.query("sex == 'male'  and  alive == 'no'"))
```

### Renamed Column 'class' to 'ticketType'

```python
df = df.rename(columns={'class': 'ticketType'})
```

#### Survivors based on passenger class

```python
a, b, c = len(df.query("ticketType == 'First' and alive == 'yes'")), len(df.query("ticketType == 'Second' and alive == 'yes'")), len(df.query("ticketType == 'Third' and alive == 'yes'"))
print("First:",a,"\nSecond: ",b,"\nThird: ",c)
```

#### Relationship between fare and survival

df.value_counts()

#### Embark towns with highest values of survivals

```python
print("Fare less than 10 but survived: ",len(df.query('embark_town <= "" and survived == 1')),
"\nFare less than 10 but dead: ", len(df.query('embark_town <= "Southampton" and survived == 1')),
"\nFare less than 10 alive females: ", len(df.query('fare <= 10 and survived == 1')),
)
```

#### Count of females who have survived P class = third

```python
print("female alive and third class: ",len(df.query("sex == 'female'  and  alive == 'yes' and ticketType == 'Third'")))
```
